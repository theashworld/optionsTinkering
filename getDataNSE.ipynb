```python
import requests

def getSymbol(symbol="NIFTY", stock=False):
    baseurl = "https://www.nseindia.com/"
    url = f"https://www.nseindia.com/api/option-chain-indices?symbol={symbol}"
    if stock:
        url = f"https://www.nseindia.com/api/option-chain-equities?symbol={symbol}"
    headers = {'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, '
                            'like Gecko) '
                            'Chrome/80.0.3987.149 Safari/537.36',
            'accept-language': 'en,gu;q=0.9,hi;q=0.8', 'accept-encoding': 'gzip, deflate, br'}
    session = requests.Session()
    request = session.get(baseurl, headers=headers, timeout=5)
    cookies = dict(request.cookies)
    response = session.get(url, headers=headers, timeout=5, cookies=cookies)
    return response.json()
```

```python
#data: {'strikePrice': 7500, 'expiryDate': '29-Dec-2022', 'PE': {'strikePrice': 7500, 'expiryDate': '29-Dec-2022', 'underlying': 'NIFTY', 'identifier': 'OPTIDXNIFTY29-12-2022PE7500.00', 'openInterest': 21, 'changeinOpenInterest': 0, 'pchangeinOpenInterest': 0, 'totalTradedVolume': 0, 'impliedVolatility': 0, 'lastPrice': 0, 'change': 0, 'pChange': 0, 'totalBuyQuantity': 4600, 'totalSellQuantity': 0, 'bidQty': 4600, 'bidprice': 3.35, 'askQty': 0, 'askPrice': 0, 'underlyingValue': 17162.65}}


def analyze(resp, expiry="near", stdev=5/100):
    """
    :param resp: response from getSymbol()
    :param expiry: "near", "mid" or "far"
    :param stdev: standard deviation in percentage for bs puts
    """
    data = resp['records']['data']
    underlying = None
    symbol = None
    if "PE" not in data[0]:
        underlying = data[0]['CE']['underlyingValue']
        symbol =  data[0]['CE']['underlying']
    else:
        underlying = data[0]['PE']['underlyingValue']
        symbol =  data[0]['PE']['underlying']
    switch = {'near': 0, 'mid': 1, 'far': 2}
    expiryDate = resp["records"]["expiryDates"][switch[expiry]]

    # Find one strike above and below
    strikeBelow = 0
    strikeBelowPrice = 0
    strikeAbove = None
    bsput = 0

    for i in data:
        if i["strikePrice"] * (1+stdev) < underlying:
            bsput = i["strikePrice"]
        if i['strikePrice'] < underlying:
            strikeBelow = i["strikePrice"]
        else:
            strikeAbove = i["strikePrice"]
            break
    #print(underlying, strikeBelow, strikeAbove, bsput, IV)
    print("Underlying at ", underlying, " BS Put at ", bsput, " expiry ", expiryDate)
    bsputData = [x for x in data if x["strikePrice"] == bsput and x["expiryDate"] == expiryDate][0]["PE"]
    bsputPrice = (bsputData["bidprice"] + bsputData["askPrice"])/2
    atmSellPrice = bsputPrice * 10
    strikeBelowPrice = [x for x in data if x["strikePrice"] == strikeBelow and x["expiryDate"] == expiryDate][0]["PE"]
    print("Buy BS Put at ", bsputPrice, "Sell ATM at ", atmSellPrice,
    "current price ", strikeBelowPrice['lastPrice'], " IV ", strikeBelowPrice['impliedVolatility'])
    return bsputPrice, atmSellPrice, strikeBelowPrice['lastPrice'], strikeBelowPrice['impliedVolatility'], underlying, bsput, expiry, symbol

#resp = getSymbol("INFY", stock=True)
#analyze(resp, expiry="near")
```

```python
allDone = []
notDone = []
for s in symbols:
    try:
        resp = getSymbol(s, stock=True)
        if "records" not in resp:
            print("Please retry ", s)
            notDone.append(s)
        else:
            print(s)
            allDone.append(analyze(resp, expiry="near", stdev=8/100))
    except:
        print("Error occurred", s)
        notDone.append(s)
notDone2 = []
for s in notDone:
    try:
        resp = getSymbol(s, stock=True)
        if "records" not in resp:
            print("Please retry ", s)
            notDone2.append(s)
        else:
            print(s)
            allDone.append(analyze(resp, expiry="near", stdev=8/100))
    except:
        print("Error occurred", s)
        notDone2.append(s)


for s in ["NIFTY", "BANKNIFTY"]:
    try:
        resp = getSymbol(s, stock=False)
        if "records" not in resp:
            print("Please retry ", s)
        else:
            print(s)
            allDone.append(analyze(resp, expiry="near", stdev=8/100))
    except:
        print("Error occurred", s)
```
